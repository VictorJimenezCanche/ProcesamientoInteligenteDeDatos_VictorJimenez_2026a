```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
celsius = np.array([-40,-10,0,8,15,22,38], dtype=float)
fahrenheit = np.array([-40,14,32,46,59,72,100],dtype= float)
```

```python
plt.scatter(celsius, fahrenheit)

plt.title('CorrelaciГіn entre Celsius y Fahrenheit')
plt.xlabel('Celsisu')
plt.ylabel('Fahrenheit')

plt.show()
```

```python
X_train = celsius
Y_train = fahrenheit
```

```python
print("Creando el modelo")
model = tf.keras.Sequential()
model.add(tf.keras.layers.Input(shape=(1,)))
model.add(tf.keras.layers.Dense(units=1))
```

```python
print("Compilando el modelo")
model.compile(optimizer=tf.keras.optimizers.Adam(1.0), loss='mean_squared_error')
```

```python
print("Entrenando el modelo")
epochs_hist = model.fit(X_train, Y_train, epochs=100)
```

```python
print("Evaluando el modelo entrenado")
print("Keys:")
print(epochs_hist.history.keys())

plt.plot(epochs_hist.history['loss'])
plt.title('Progreso de PГ©rdida durante Entrenamiento del Modelo')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend(['Training Loss'])
plt.show()
```

```python
Temp_C = 100
Temp_C_array = np.array([[Temp_C]])
Temp_F = model.predict(Temp_C_array)

print("Temperatura de predicciГіn: " + str(Temp_F))
```

```python
Temp_F = 9/5 * Temp_C + 32
print("Temperatura de EcuaciГіn: " + str(Temp_F))
```

**1. El optimizador de algoritmo utilizado es Adam, pero existen otros como SGD y RMSProp, describe cada uno de ello y explica como pudieran cambiar los resultados de tu modelo.**

Adam: introducido en 2014, Adam (adaptive moment estimation) combina el impulso con RMSProp para ajustar la tasa de aprendizaje de cada parГЎmetro en funciГіn de sus gradientes anteriores. Las versiones posteriores de Adam agregaron un arranque en caliente, que aumenta gradualmente la tasa de aprendizaje al comenzar el entrenamiento.

El descenso de gradiente estocГЎstico (SGD) selecciona aleatoriamente un punto de datos por iteraciГіn, lo que aumenta enormemente la velocidad y los matices. Pero la alta frecuencia de su actualizaciГіn puede reducir la estabilidad. El SGD tiene muchas variantes, entre ellas, Adam, AdaGrad y RMSProp. 

RMSProp: el algoritmo RMSProp (root mean square propagation) ajusta el peso de aprendizaje para cada parГЎmetro de acuerdo con un promedio mГіvil de los cuadrados de cada gradiente. Mejora en AdaGrad al ignorar los gradientes en el pasado lejano, aumentando asГ­ la estabilidad y facilitando una convergencia mГЎs rГЎpida.

**5.Que funciГіn de activaciГіn utiliza el modelo? ВїEs posible cambiarla? Si es asГ­, cambia a una funciГіn RELU y muestra el resultado.**

El modelo utiliza por defecto una funciГіn de activaciГіn lineal, ya que no se especifica otra en las capas Dense. Esta funciГіn es adecuada para este problema porque la relaciГіn entre Celsius y Fahrenheit es lineal. Sin embargo, es posible cambiarla, por ejemplo a ReLU, al especificarlo al momento de agregar una capa.

**Usando ReLU**

```python
#Cambiando a RelU
model_relu = tf.keras.Sequential()
model_relu.add(tf.keras.layers.Input(shape=(1,)))
model_relu.add(tf.keras.layers.Dense(units=3, activation='relu'))
model_relu.add(tf.keras.layers.Dense(units=3, activation='relu'))
model_relu.add(tf.keras.layers.Dense(units=1))
```

```python
print("Compilando el modelo de tres capas")
model_relu.compile(optimizer=tf.keras.optimizers.Adam(1.0), loss='mean_squared_error')
```

```python
print("Entrenando el modelo con ReLu")
epochs_hist_relu = model_relu.fit(celsius, fahrenheit, epochs=100)
```

```python
print("Evaluando el modelo entrenado ReLu")
print("Keys:")
print(epochs_hist_relu.history.keys())

plt.plot(epochs_hist_relu.history['loss'])
plt.title('Progreso de PГ©rdida durante Entrenamiento del Modelo Relu')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend(['Training Loss'])
plt.show()
```

```python
Temp_C = 100
Temp_C_array = np.array([[Temp_C]])
Temp_F = model_relu.predict(Temp_C_array)

print("Temperatura de predicciГіn: " + str(Temp_F))
```

```python
Temp_F = 9/5 * Temp_C + 32
print("Temperatura de EcuaciГіn: " + str(Temp_F))
```

Al cambiar la funciГіn de activaciГіn a ReLU, el modelo dejГі de predecir correctamente los valores. Esto ocurre porque ReLU convierte todos los valores negativos en cero, lo que provoca pГ©rdida de informaciГіn en el conjunto de datos, ya que existen temperaturas negativas en grados Celsius. Debido a esto, el modelo no logra aprender correctamente la relaciГіn lineal entre Celsius y Fahrenheit. Por esta razГіn, la funciГіn de activaciГіn lineal resulta mГЎs adecuada para este problema de regresiГіn.
