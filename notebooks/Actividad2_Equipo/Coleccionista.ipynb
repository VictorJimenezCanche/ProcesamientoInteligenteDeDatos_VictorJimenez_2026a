# Coleccionista

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Importe de Datos
data = pd.read_csv("./Coleccionista.csv")
data.head()
```

# 1. Determinar las variables independientes

```python
X = data[['Bidders', 'Age']]
X 
```

# #2. Determinar la variable dependiente

```python
y = data[['Price']]
y
```

```python
# Ajustar el modelo de regresi칩n m칰ltiple
model = LinearRegression()
model.fit(X, y) #Entrenando el modelo
y_pred = model.predict(X) #Predicciones del modelo
y_pred

#M칠tricas estad칤sticas
mse = mean_squared_error(y, y_pred) # Error de la media cuadr치tica
r2 = r2_score(y, y_pred) # Varianza
```

# 3. Valor de Intersecci칩n

```python
model.intercept_
```

```python
# Coeficientes de la regresi칩n
print("Coeficientes: ", model.coef_)
print("Error de la media cuadr치tica (MSE): ", mse)
print("Varianza (R^2): ", r2)
```

# 4. Estimaci칩n de Precio de Subasta

```python
Y = model.predict(np.array([[14,80]]))
print("Precio estimado de una subasta: ", Y)
```
