# Regresión Lineal Múltiple - SalsBerry Realty
1. Determinar las variables independientes
2. Determinar la variable dependiente
3. Obtener el valor de intersección
4. Estimar el corto de calefacción para una Temperatura externa media de 30°, 5 pulgadas de aislamiento y un calentador con antigüedad de 10 años

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

#Importe de Datos
df = pd.read_csv("./SalsberryRealty.csv")
df.head()
```

# 1. Determinar las variables independientes

```python
X = df[['Temp', 'Insul', 'Age']]
X
```

# #2. Determinar la variable dependiente

```python
y =df[['Cost']]
y
```

```python
# Ajustar el modelo de regresión múltiple
model = LinearRegression()
model.fit(X, y) #Entrenando el modelo
y_pred = model.predict(X) #Predicciones del modelo
y_pred

#Métricas estadísticas
mse = mean_squared_error(y, y_pred) # Error de la media cuadrática
r2 = r2_score(y, y_pred) # Varianza
```

# 3. Valor de Intersección

```python
model.intercept_
```

```python
# Coeficientes de la regresión
print("Coeficientes: ", model.coef_)
print("Error de la media cuadrática (MSE): ", mse)
print("Varianza (R^2): ", r2)
```

# 4. Estimar para 30°, aislamiento 5",antigüedad 10 años

```python
Y = model.predict(np.array([[30, 5, 10]]))
print("El costo de calefacción estimado es: ", Y)
```
