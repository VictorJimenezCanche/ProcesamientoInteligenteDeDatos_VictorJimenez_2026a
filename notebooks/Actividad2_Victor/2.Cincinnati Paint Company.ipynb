Regresión Lineal Múltiple - Cincinnati Paint Company
1. Determinar las varibles independientes
2. Determinar la variable dependiente
3. Obtener el valor de intersección
4. Estimar la comisión

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Importar datos
df = pd.read_csv('../datasets/CincinnatiPaintCompany.csv')
df.head()
```

# 1. Determinar las variables independientes
   x1,x2

```python
X=df[['Driven','Calls']]
X
```

# 2. Determinar la variable dependiente Y

```python
y=df[['Commissions']]
y
```

```python
#Ajustar el modelo de regresión múltiple
model = LinearRegression()
model.fit(X,y) #entrenar el modelo
y_pred = model.predict(X)
#y_pred

#Métricas estadísticas
mse = mean_squared_error(y,y_pred) #Error de la media cuadrática
r2 = r2_score(y, y_pred) #varianza
```

# 3. Valor de intersección

```python
model.intercept_
```

```python
#coeficientes
print('Coeficientes: ',model.coef_ )
print('Error cuadratico de la media: ',mse)
print('Varianza: ',r2)
```

# 4. Estimar las comisiones

```python
Y = model.predict([[139,2371]])
print("La comisión es: ",Y)
```
