Regresión Lineal Múltiple - SalsBerry Realty
1. Determinar las varibles independientes
2. Determinar la variable dependiente
3. Obtener el valor de intersección
4. Estimar el costo de calefacción para una temperatura externa media de 30°F, 5 pulgadas de aislamiento y un calentador con antigüedad de 10 años.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Importar datos
df = pd.read_csv('../datasets/SalsberryRealty.csv')
df.head()
```

# 1. Determinar las variables independientes
   x1,x2,x3

```python
X=df[['Temp','Insul','Age']]
X
```

# 2. Determinar la variable dependiente Y

```python
y=df[['Cost']]
y
```

```python
#Ajustar el modelo de regresión múltiple
model = LinearRegression()
model.fit(X,y) #entrenar el modelo
y_pred = model.predict(X)
#y_pred

#Métricas estadísticas
mse = mean_squared_error(y,y_pred) #Error de la media cuadrática
r2 = r2_score(y, y_pred) #varianza
```

# 3. Valor de intersección

```python
model.intercept_
```

```python
#coeficientes
print('Coeficientes: ',model.coef_ )
print('Error cuadratico de la media: ',mse)
print('Varianza: ',r2)
```

# 4. Estimar para 30º, aislamiento de 5¨, antiguedad 10 años

```python
Y = model.predict([[30,5,10]])
print("El costo de calefacción estimado es: ",Y)
```
