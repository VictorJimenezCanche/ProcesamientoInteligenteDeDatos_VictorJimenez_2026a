Regresión Lineal Múltiple - High Point Sofa and Chair
1. Determinar las varibles independientes
2. Determinar la variable dependiente
3. Obtener el valor de intersección
4. Estimar el desempeño laboral

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Importar datos
df = pd.read_csv('../datasets/HighPointSofaandChair.csv')
df.head()
```

# 1. Determinar las variables independientes
   x1,x2

```python
X=df[['Aptitude','Union']]
X
```

# 2. Determinar la variable dependiente Y

```python
y=df[['Performance']]
y
```

```python
#Ajustar el modelo de regresión múltiple
model = LinearRegression()
model.fit(X,y) #entrenar el modelo
y_pred = model.predict(X)
#y_pred

#Métricas estadísticas
mse = mean_squared_error(y,y_pred) #Error de la media cuadrática
r2 = r2_score(y, y_pred) #varianza
```

# 3. Valor de intersección

```python
model.intercept_
```

```python
#coeficientes
print('Coeficientes: ',model.coef_ )
print('Error cuadratico de la media: ',mse)
print('Varianza: ',r2)
```

# 4. Estimar el desempeño laboral

```python
Y = model.predict([[5,0]])
print("El desempeño laboral con una aptitud: ",Y)
```
