*1.2 UN POCO DE BIOLOGÍA*

Esta sección da un repaso a los conceptos bilogicos sobre los cuales se trabajan y se intentan simular.
Concepto: Fenotipo, entendio como aquellas caracteristicas distingibles de un organismo, visibles o no(como color de cabello, color de ojos, tipo de sangre, etc)
Es el producto del ambiente donde se desarrolla el individuo y la herencia dada.
Concepto: Celula somática. Celula que conforma el organismo. Posee dos juegos de cromosomas, uno dado por la madre y el otro por le padre. En total son 23.
Concepto: Cromosoma. Molecula de ADN compuesta de 4 bases: Adenina, guanina, citosina y timina.
Concepto: Proteína. Cadena de 3 nucleotidos que conforman un aminoacido que se unen con otros aminoacidos producidos por otros tercetos.
Concepto: Codones. Cadena de 3 nucleotidos.
Concepto: Genes. Conjunto de nucleotidos que codifican una proteina completa.
Concepto: Genoma. Conjunto de todos los cromosomas de un individuo.
Concepto: Genotipo: Conjunto de genes contenido en el genoma. Determina en gran medida el fenotipo del individuo.
Concepto: Gametos. Se reproducen por meiosis. Producen celulas haploides.

Pasos de la meiosis.
1. Aparece la celula diploide original. Se duplica el numéro de cromosomas en la célula.
2. Se cruza un juego de cromosomas de la madre con el del padre, formando uno hibrido.
3. Se devide la celula y quedan 4 pares. Uno puro de madre y padre y dos hibridos.

En el paso dos se hace una ruptura y unión llamados *quiasmas* que generan segmentos llamados *cromátidas*
En el paso uno, donde se duplican los cromosomas, entra en función la encima ADN-Polimerasa, pero pueden haber errores en ese proceso. Estos errores son las mutaciones, si bien en general tienden a ser desfavorables, algunos ayudan o no presentan problemas al individuo, permitiendo que se puedan heredar.

**2. Clases**

```python
import random

# Clase Individuo
class Individuo:
    def __init__(self, genes):
        self.genes = genes  # Lista de valores (ej: binarios)
        self.fitness = self.calcular_fitness()

    def calcular_fitness(self):
        # Ejemplo simple: suma de genes
        return sum(self.genes)

    def mutar(self, probabilidad=0.1):
        for i in range(len(self.genes)):
            if random.random() < probabilidad:
                self.genes[i] = 1 - self.genes[i]  # Cambia 0 por 1
        self.fitness = self.calcular_fitness()

    def __repr__(self):
        return f"Individuo({self.genes}, fitness={self.fitness})"


# Clase Poblacion
class Poblacion:
    def __init__(self, tamaño, longitud_gen):
        self.individuos = [
            Individuo([random.randint(0, 1) for _ in range(longitud_gen)])
            for _ in range(tamaño)
        ]

    def evaluar(self):
        for individuo in self.individuos:
            individuo.fitness = individuo.calcular_fitness()

    def seleccionar_mejores(self, n):
        return sorted(self.individuos, key=lambda x: x.fitness, reverse=True)[:n]

    def cruzar(self, padre1, padre2):
        punto = random.randint(1, len(padre1.genes) - 1)
        hijo_genes = padre1.genes[:punto] + padre2.genes[punto:]
        return Individuo(hijo_genes)

    def nueva_generacion(self):
        nuevos = []
        mejores = self.seleccionar_mejores(2)

        while len(nuevos) < len(self.individuos):
            hijo = self.cruzar(mejores[0], mejores[1])
            hijo.mutar()
            nuevos.append(hijo)

        self.individuos = nuevos

    def __repr__(self):
        return f"Poblacion({self.individuos})"
```
