```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
```

```python
comisiones = [22,13,33,38,23,47,29,38,41,32,20,13,47,
              38,44,29,38,37,14,34,25,27,25,43,34]

llamadas = [139,132,144,142,142,142,138,139,144,134,135,137,146,
            146,144,147,144,149,131,144,132,132,127,154,147]

millas = [2371,2226,2731,3351,2289,3449,3114,3342,2842,2625,2121,2219,3463,
          3290,3103,2122,2791,3209,2287,2848,2690,2933,2671,2988,2829]

df = pd.DataFrame({
    "Comisiones": comisiones,
    "Llamadas": llamadas,
    "Millas": millas
})
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df["Llamadas"], df["Millas"], df["Comisiones"])

ax.set_xlabel("Llamadas")
ax.set_ylabel("Millas Recorridas")
ax.set_zlabel("Comisiones")
ax.set_title("Correlacin Mltiple")

plt.show()
```

```python
print("Correlacin Comisiones-Llamadas:",
      round(df["Comisiones"].corr(df["Llamadas"]),4))

print("Correlacin Comisiones-Millas:",
      round(df["Comisiones"].corr(df["Millas"]),4))
```

```python
X = df[["Llamadas","Millas"]]
Y = df["Comisiones"]

modelo = LinearRegression()
modelo.fit(X, Y)

R_cuadrado = modelo.score(X, Y)

print("\nCoeficiente de determinacin mltiple R:",
      round(R_cuadrado,4))
```

**Conclusin**
Existe una correlacin positiva, ya que se observa una ligera tendencia: a mayor ocupacin, mayor ingreso.
